```python
#%%

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
#%%

bert_results = ['NEGATIVE',
 'POSITIVE',
 'NEGATIVE',
 'POSITIVE',
 'POSITIVE',
 'NEGATIVE',
 'POSITIVE']

llm_results = ['negative',
 'positive',
 'positive',
 'positive',
 'positive',
 'positive',
 'positive']
```

```python
#%%

# Creamos un DataFrame donde las filas son transcripciones y las columnas son modelos
def label_to_numeric(label):
    return 1 if label.lower() == "positive" else 0

BERT_numeric = [label_to_numeric(label) for label in bert_results]
LLM_numeric = [label_to_numeric(label) for label in llm_results]

df = pd.DataFrame({
    "BERT": BERT_numeric,
    "FLAN-T5": LLM_numeric
}, index=range(1,len(LLM_numeric)+1))

# Definimos un mapa de colores donde 0 (negativo) = rojo y 1 (positivo) = verde
cmap = ListedColormap(["#F08080", "#98FB98"])

sns.set_theme(style="whitegrid", font_scale=1.1)

# Graficamos el heatmap
plt.figure(figsize=(6, 3))  # tamaño de la figura opcional
sns.heatmap(
    df,
    cmap=cmap, 
    cbar=False,
    linewidths=2,
    linecolor="white"
)

# Ajustes de ejes y título
plt.title("Comparación de clasificación")
plt.ylabel("Transcript")
plt.xlabel("Modelo")
plt.xticks(rotation=0)

plt.show()
```
